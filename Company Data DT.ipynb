```python
comp = pd.get_dummies(company,columns=['ShelveLoc','Urban','US'],drop_first=True)
```

```python
conditions=[
    (comp['Sales']>=10),
    (comp['Sales']<=10) & (comp['Sales']>=5),
    (comp['Sales']<5)
]
values = ['High','Medium','Low']
comp['Sales']=np.select(conditions,values)
```

```python
comp
```

```python
label_encoder = preprocessing.LabelEncoder()
comp['Sales']= label_encoder.fit_transform(comp['Sales']) 
```

```python
x = comp.iloc[:,1:11]
y = comp.iloc[:,0]
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=40)
```

```python
model = DecisionTreeClassifier(criterion='entropy',max_depth=5)
model.fit(x_train,y_train)
```

```python
tree.plot_tree(model);
```

```python
model.score(x_train,y_train)
```

```python
model = DecisionTreeClassifier(criterion='entropy')
param_grid = [{'max_depth':[2,3,4,5,6],'max_features':[3,4,5,6,7,8,9],'max_leaf_nodes':[8,10,12,15,18] }]
gsv = GridSearchCV(model,param_grid,cv=10)
gsv.fit(x_train,y_train)
```

```python
gsv.score(x_train,y_train)
```

```python
y_pred = gsv.predict(x_test)
```
